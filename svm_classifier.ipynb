```r
# import neccessary libraries
library(e1071)
```

```r
# load feature files
train.glove <- read.csv('./train_glove_final.csv')
test.glove <- read.csv('./test_glove_final.csv')

dim(train.glove)
dim(test.glove)
```

```r
# shuffle data
train.glove <- train.glove[sample(nrow(train.glove)),]
dim(train.glove)
```

```r
head(train.glove)
```

```r
head(test.glove)
```

```r
# build model from train data
svm.model <- svm(label ~., data = train.glove, kernel='radial', gamma = 0.001, cost=3.162)
```

```r
# predict test labels
predictions <- predict(svm.model, test.glove)
predictions
```

```r
test.doc <- read.csv('./test.csv')
```

```r
output <- cbind(test.doc, predictions)[,-2]
```

```r
dim(output)
```

```r
# output text file
write.table(output, file = "testing_labels_pred.txt", sep = " ", quote = FALSE, row.names = FALSE, col.names=F)
```
